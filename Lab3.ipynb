```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('/content/lab2/canada_per_capita_income.csv')

# Split features and target
X = df[['year']]
y = df['per capita income (US$)']

# Build model
model = LinearRegression()
model.fit(X, y)

# Predict income for 2020
prediction = model.predict([[2020]])
print("Predicted income in 2020:", prediction[0])

# Visualization
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Year")
plt.ylabel("Income")
plt.title("Canada Income Prediction")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/lab2/salary.csv')

# Handle missing values
df['YearsExperience'] = df['YearsExperience'].fillna(df['YearsExperience'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

X = df[['YearsExperience']]
y = df['Salary']

model = LinearRegression()
model.fit(X, y)

salary_12 = model.predict([[12]])
print("Predicted salary:", salary_12[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/lab2/hiring.csv')

# Convert word numbers to digits
word_to_num = {
    "zero":0,"one":1,"two":2,"three":3,"four":4,
    "five":5,"six":6,"seven":7,"eight":8,"nine":9,
    "ten":10,"eleven":11,"twelve":12
}

df['experience'] = df['experience'].replace(word_to_num)

# Fill missing values (safe way)
df['experience'] = df['experience'].fillna(0)
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].mean()
)

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

model = LinearRegression()
model.fit(X, y)

print(model.predict([[2,9,6]]))
print(model.predict([[12,10,10]]))
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/lab2/1000_Companies.csv')

# One hot encoding
df = pd.get_dummies(df, columns=['State'], drop_first=True)

X = df.drop('Profit', axis=1)
y = df['Profit']

model = LinearRegression()
model.fit(X, y)

# Prediction example
test_data = [[91694.48, 515841.3, 11931.24, 0, 1]]  # encoded Florida
print(model.predict(test_data))
```
